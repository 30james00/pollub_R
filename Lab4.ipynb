Możesz np. użyć polecenia z w/w paczki: str_sub(..string phrase.., ..start position.., ..end position..),
które zwraca ze stringa jego podstring o zadanych parametrach.
Rozszerzamy ramkę table_ o pożądaną kolumnę:
`tableWithFirstCipherColumn = data.frame(table_, ‘C1’=firstCipher_(table_$Ludność))`

````r
library(stringr)
first_cipher_ <- function(n) {
  return (str_sub(n, 1, 1))
}
second_cipher_ <- function(n) {
  return (str_sub(n, 2, 2))
}
third_cipher_ <- function(n) {
  return (str_sub(n, 3, 3))
}
table_$C1 = first_cipher_(table_$Ludnosc)
table_
````
